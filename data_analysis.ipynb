```python
import pandas as pd

data = pd.read_csv(r"C:\Users\AK\Downloads\Data_Analytics_Project\customer_shopping_behavior.csv")

data.head()
```

```python
data.info()
```

```python
data.describe(include="all")
```

```python
data.isnull().sum()
```

```python
data.groupby ("Category")["Purchase Amount (USD)"].sum()
```

```python
data["Review Rating"]= data.groupby('Category')["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
data.columns= data.columns.str.lower()
data.columns= data.columns.str.replace(' ', '_')
data=data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
data.columns
```

```python
#Age group column
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
data['age_group'] = pd.qcut(data['age'], q=4, labels=labels)
```

```python
data[['age','age_group']].head(10)
```

```python
#purchase frequecny
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90,
}
data['purchase_frequency_days'] = data['frequency_of_purchases'].map(frequency_mapping)
```

```python
data[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
data[['discount_applied', 'promo_code_used']].head(10)
```

```python
(data['discount_applied']==data['promo_code_used']).all()
```

```python
data=data.drop('promo_code_used',axis=1)
```
